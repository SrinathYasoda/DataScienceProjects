## Question 5: Train Test splits

```python

X=data.drop(['Yearly_Amount_Spent','Email','Address','Avatar'],axis=1)
Y=data['Yearly_Amount_Spent']

X_train,X_test,Y_train,Y_test=sklearn.model_selection.train_test_split(X,Y,test_size=0.20,random_state=85)

lm=LinearRegression()
lm.fit(X_train,Y_train) 
```

## R squre value

```python
lm.score(X_train,Y_train)
```

## Coefficient and Variance

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
#print(lm.predict(X_test))
Y_train_pred=lm.predict(X_train)
Y_test_pred=lm.predict(X_test)
```

```python
df=pd.DataFrame(Y_train_pred,Y_train)
print(df)
```

```python
mse_test=mean_squared_error(Y_test,Y_test_pred)
print('testing pred',mse_test)
```

```python
mse_train=mean_squared_error(Y_train,Y_train_pred)
print('training pred',mse_train)
```

```python
RMS_train=math.sqrt(mse_train)
RMS_test=math.sqrt(mse_test)
print(RMS_train)
print(RMS_test)
```

## Point 6: Scatter plot between Acutal and Predicted Values


```python
plt.scatter(Y_train_pred, Y_train_pred - Y_train,c='blue',marker='o',label='Training data')
plt.scatter(Y_test_pred, Y_test_pred - Y_test,c='lightgreen',marker='s',label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc= 'upper left')
plt.hlines(y=0,xmin=0,xmax=1000)
plt.plot()
plt.show()
```
